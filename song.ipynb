```python
stemmer = PorterStemmer()
```

```python
def token(txt):

  tokens = nltk.word_tokenize(txt)
  a = [stemmer.stem(i) for i in tokens]
  b = " ".join(a)
  return b
```

```python
token("you are beautiful, beauty")
```

```python
df1['text'].apply(lambda x: token(x))
```

```python
tfid = TfidfVectorizer(analyzer = 'word',stop_words = 'english')
```

```python
matrix = tfid.fit_transform(df['text'])
```

```python
similarity = cosine_similarity(matrix)
```

```python
similarity[0]
```

```python
df1[df1['song'] == 'Waiting For The Man'].index[0]
```

```python
def recommend(song_name):
  idx = df1[df1['song'] == song_name].index[0]
  distance = sorted(list(enumerate(similarity[idx])),reverse = True,key = lambda x:x[1])
  song = []
  for i in distance[1:5]:
    song.append(df1.iloc[i[0]].song)
  return song
```

```python
recommend("In Your Eyes")
```
